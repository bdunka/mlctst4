```python
import os
import subprocess
import speech_recognition as sr
import io
import joblib
from pydub import AudioSegment
from pytube import YouTube
from google.cloud import storage
from google.cloud import speech
from textblob import TextBlob
from google.cloud import speech_v1p1beta1 as speech
from google.cloud import speech_v1p1beta1 as speech
from google.cloud.speech_v1p1beta1 import types

#Specify path to store transcript file
path = "C:\\Users\\kumar\\Path_to_store_transcript_file\\video"

#Download video
def video_download(url):
    yt = YouTube(url).streams.first().download(path)

#Convert video to .flac format with mono channel
def video_convert():
    mylist = os.listdir(path)
    for filename in mylist:
        if "flac" not in filename:
            print("file found with no flac")
            f = os.path.splitext(filename)[0]
            print(f)
            fin = f + ".3gpp"
            print(fin)
            fout = f + ".flac"
            print(fout)
            os.chdir(path)
            print(os.getcwd())
            subprocess.call(['ffmpeg', '-i', fin, '-c:a', 'flac', fout])
            subprocess.call(['ffmpeg', '-i', fout,'-ac', '1', f+"_mono.flac"])
#             os.remove(fin)
    return f+"_mono.flac"

#Get Sentiment Analysis
def sentiment_analysis(result):
        model=joblib.load(r'C:\Users\kumar\Path_to_store_transcript_file\model.pkl')
        cv=joblib.load(r'C:\Users\kumar\Path_to_store_transcript_file\vectorizer.pkl')
        inp = cv.transform([result]).toarray()
        return model.predict(inp)

#Download video from given url
url = input("Enter the url: ")
print("Video Download Begins...")
video_download(url)
print ("Video Downloaded!")

#Convert to flac format with mono channel
audio_file = video_convert()
print ("Conversion complete!!")
#convert .flac to .wav

# flac_path = r"C:\Users\kumar\Path_to_store_transcript_file\Israel - Birth of a state  DW Documentary.flac"
# wav_path = r"C:\Users\kumar\Path_to_store_transcript_file\Israel - Birth of a state  DW Documentary.wav"

a=os.listdir(path)[0]
b=a.split('.')[0]+'.flac'
w=a.split('.')[0]+'.wav'
flac_path=path+'\\'+b
wav_path=path+'\\'+w

flac = AudioSegment.from_file(flac_path, format='flac')
flac.export(wav_path, format='wav')

def audio_to_text(wav_path):
    # Initialize the recognizer
    r = sr.Recognizer()

    # Load the audio file
    with sr.AudioFile(wav_path) as source:
        # Read the entire audio file
        audio_data = r.record(source)

        # Convert speech to text
        text = r.recognize_google(audio_data)

    # Return the converted text
    return text


# file_path = r"C:\Users\kumar\Path_to_store_transcript_file\Improve English Speaking Skills Everyday (Tips to speak in English) English Conversation Practice.wav"
#file_path=wav_path
result = audio_to_text(wav_path)
with open (r'C:\Users\kumar\Path_to_store_transcript_file\video\transcript.txt','w') as f:
    f.write(result)
    
    
#Sentiment of transcript
sentiment_value = sentiment_analysis(result)
print("Sentiment Analysis: "+ str(sentiment_value))
```
